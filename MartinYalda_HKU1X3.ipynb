Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House:
  def __init__(self, area, num_rooms, num_bathrooms, num_floors, has_garage, has_garden, house_type="Brick", location="Budapest"):
    self.area = area
    self.num_rooms = num_rooms
    self.num_bathrooms = num_bathrooms
    self.num_floors = num_floors
    self.has_garage = has_garage
    self.has_garden = has_garden
    self.house_type = house_type
    self.location = location

  def price_calculation(self):
    house_price = self.area * 2000

    if self.has_garage:
      house_price += 20000

    if self.has_garden:
      house_price += 30000

    return house_price
```

###A) (5 points)
 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

which methods belong to parent


which should be overridden


which attributes are shared. Explain your choice.

```python
class Animal:
  def __init__(self, name, age, color):
    self.name = name
    self.age = age
    self.color = color

  def make_sound(self):
    print("This animal makes unknown sound.")
```

```python
class Dog(Animal):
  def __init__(self, name, age, color, is_trained):
    super().__init__(name, age, color)
    self.is_trained = is_trained

  def make_sound(self):
    print(f"{self.name} says: Woof Woof!")

  def catch_ball(self):
    print(f"{self.name} is chasing a tennis ball!")


class Cat(Animal):
  def __init__(self, name, age, color, is_indoor):
    super().__init__(name, age, color)
    self.is_indoor = is_indoor

  def make_sound(self):
    print(f"{self.name} says: Meow!")

  def take_nap(self):
    print(f"{self.name} is sleeping in a sunny spot.")


class Bird(Animal):
  def __init__(self, name, age, color, can_fly):
    super().__init__(name, age, color)
    self.can_fly = can_fly

  def make_sound(self):
    print(f"{self.name} says: Tweet Tweet!")

  def build_nest(self):
    print(f"{self.name} is gathering twings for a nest.")
```

```python
dog = Dog("Max", 3, "Brown", True)
cat = Cat("Bella", 2, "Grey", True)
bird = Bird("Rocky", 1, "Blue", True)

print(f"{dog.name}'s Profile")
print(f"Age: {dog.age}, Color: {dog.color}, Trained: {dog.is_trained}")
dog.make_sound()
dog.catch_ball()
print("-" * 35)

print(f"{cat.name}'s Profile")
print(f"Age: {cat.age}, Color: {cat.color}, Indoor: {cat.is_indoor}")
cat.make_sound()
cat.take_nap()
print("-" * 35)

print(f"{bird.name}'s Profile")
print(f"Age: {bird.age}, Color: {bird.color}, Can Fly: {bird.can_fly}")
bird.make_sound()
bird.build_nest()
```

**Methods belonging to parent:** I added make_sound to the parent class because making noise is a general behavior that applies to all animals.

**Methods that should be overridden:** I chose to override make_sound inside Dog, Cat, and Bird because they don't make the same noise. A dog barks, a cat meows, and a bird tweets, so each subclass needs its own print statement.

**Attributes that are shared:** I put name, age, and color in the Animal class since every animal has a name, an age, and a color. This keeps the code clean so I don't have to type them out for every single subclass.

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
class Shape:
  def area(self):
    pass
```

```python
import math

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return math.pi * (self.radius ** 2)


class Square(Shape):
  def __init__(self, side):
    self.side = side

  def area(self):
    return self.side * self.side
```

```python
circle = Circle(5)
square = Square(4)

print(f"Circler Area: {round(circle.area(), 2)}")
print(f"Square Arera: {square.area()}")
```

```python
shapes = [circle, square]

for s in shapes:
  print(round(s.area(), 2))
```

Create a list containing only the vowels from all words.

```python
words = ["apple", "sky", "education", "why"]
```

Expected output: ['a', 'e', 'e', 'u', 'a', 'i', 'o']

```python
vowels_list =[]

vowels = "aeuio"

for word in words:
  for letter in word:
    if letter in vowels:
      vowels_list.append(letter)

print(vowels_list)
```

Combine two lists into tuples

```python
names = ["Anna", "Bob", "Charlie"]
scores = [85, 92, 78]
```

Expected output: [('Anna', 85), ('Bob', 92), ('Charlie', 78)]

```python
combined_list = []

for i in range(len(names)):
  pair = names[i], scores[i]
  combined_list.append(pair)

print(combined_list)
```

Create a dictionary from two lists using the two lists from above (names, scores)

Expected output:
{
    "Anna": 85,
    "Bob": 92,
    "Charlie": 78
}

```python
scores_dict = {}

for i in range(len(names)):
  scores_dict[names[i]] = scores[i]

print(scores_dict)
```

Count character frequency

```python
text = "banana"
```

Expected output: {'b': 1, 'a': 3, 'n': 2}

```python
char_count = {}

for char in text:
  if char in char_count:
    char_count[char] += 1
  else:
    char_count[char] = 1
print(char_count)
```
